```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'hospital_id': [1,2,3,4,5],
    'num_camas': [100,120,110,np.nan,140],
    'indice_satisfaccion': [7.5, 8.0, np.nan, 7.0, 8.5]
})
```

```python
print("--- Mean ---")
print("Pandas mean:", df['indice_satisfaccion'].mean())
print("Numpy mean:", np.mean(df['indice_satisfaccion']))
print("Numpy nanmean:", np.nanmean(df['indice_satisfaccion']))

print("--- Sum ---")
print("Pandas var:", df['num_camas'].sum())
print("Numpy var:", np.sum(df['num_camas']))

print("--- Max ---")
print("Pandas var:", df['num_camas'].max())
print("Numpy var:", np.max(df['num_camas']))

print("--- Min ---")
print("Pandas var:", df['num_camas'].min())
print("Numpy var:", np.min(df['num_camas']))

print("--- Prod ---")
print("Pandas var:", df['num_camas'].prod())
print("Numpy var:", np.prod(df['num_camas']))

print("--- Size ---")
print("Pandas var:", df['num_camas'].size)
print("Numpy var:", np.size(df['num_camas']))

print("--- Quantile/Percentile ---")
print("Pandas var:", df['indice_satisfaccion'].quantile(0.25))
print("Numpy var:", np.percentile(df['indice_satisfaccion'],25))

print("--- Std ---")
print("Pandas var:", df['num_camas'].std())
print("Numpy var:", np.std(df['num_camas']))
print("Numpy var ddof=1:", np.std(df['num_camas'], ddof=1))

print("--- Median ---")
print("Pandas var:", df['num_camas'].median())
print("Numpy var:", np.median(df['num_camas']))

print("--- Var ---")
print("Pandas var:", df['num_camas'].var())
print("Numpy var:", np.var(df['num_camas']))
print("Numpy var ddof=1:", np.var(df['num_camas'], ddof=1))
```

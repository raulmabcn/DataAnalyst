```python
import pandas as pd


def p( to_print ):
  print( to_print )
  print("------------------")
```

## Crear Series y DataFrames

```python
s = pd.Series([100,200,300,400], index=['A','B','C','D'])
data = {'Name':['Alice','Bob','Charlie'],'Age':[25,30,35]}
df = pd.DataFrame(data)

p(s)
p(df)
```

## Creación y Manipulación

```python
s1 = pd.Series([10,20,30,40,50])
# crear serie con indice personalizado
s2 = pd.Series([100,200,300,400], index=['a','b','c','d'])
# crear una serie desde un diccionario
data_dict = {'x':1,'y':2,'z':3}
s3 = pd.Series(data_dict)
# crear serie desde un valor escalar
s4 = pd.Series(99, index=['a','b','c','d'])

p(s1)
p(s2)
p(s3)
p(s4)
```

## Accediendo elementos en las series

```python
# accediendo por indice entero
p(s1.iloc[0])  #primer elemento de la serie
p(s1.iloc[-1]) #ultimo elemento de la serie
p(s1.iloc[1:3]) #desde el segundo elemento hasta el tercero ( el indice 3 corresponde con el cuarto elemento)

p(s2.loc['a']) # valor del indice: a
p(s2.loc['b':'d']) # valores entre los indices b y c
```

## Filtrar datos en Series

```python
p(s1[s1 > 20])
```

## DataFrame Creación y Manipulación

```python
data = {
    'Name'  : ['Alice','Bob','Charlie'],
    'Age'   : [25, 30, 35],
    'Salary': [50000,60000,70000]
}
df1 = pd.DataFrame(data)
data_list = [['Alice',25,50000],['Bob',30,60000],['Charlie',35,70000]]
df2 = pd.DataFrame(data_list,columns=['Name','Age','Salary'])

p(df1)
p(df2)
```

## Indexando DataFrames

```python
df1.set_index('Name',inplace=True)
p(df1)
df1.reset_index(inplace=True)
p(df2)
```

## Seleccionando Datos

```python
p(df1['Age'])             # Selecciona Valores de una columna
p(df1[['Name','Salary']]) # Selecciona Valores de dos columnas
p(df1.iloc[1])            # Selecciona el valor de la posición del indice 1
p(df1.iloc[0:2])          # Selecciona los valores entre el indice 0 y
df1.set_index('Name', inplace=True)
p(df1.loc['Alice'])
```

## Seleccion de Datos y Filtrado

```python
data = {
    'Name'  : ['Alice','Bob','Charlie','David'],
    'Age'   : [25, 30, 35, 40],
    'Salary': [50000,60000,70000,80000]
}
df = pd.DataFrame(data)
age_series = df['Age'] #Devuelve una sola columna (una Serie)
p(age_series)

name_salary_df = df[['Name','Salary']] #Devuelve dos columnas (un DataFrame )
p(name_salary_df)

non_existent = df.get('InvalidColumn', 'Column not found')
p(non_existent)
```

## Selección de Filas

```python
row_1  = df.loc[1]
p(row_1)
rows_1_2 = df.loc[1:2]
p(rows_1_2)
rows_specific = df.loc[1:3,['Name','Salary']]
p(rows_specific)
row_0 = df.iloc[0]
p(row_0)
rows_0_2 = df.iloc[0:2]
p(rows_0_2)
rows_cols_specific = df.iloc[1:3,[0,2]]
p(rows_cols_specific)
filtered_df = df.query('Age > 30')
```

## Filtro Condicionado

```python
filtered_age = df[df['Age'] > 30 ]
p(filtered_age)
filtered_query = df.query('Salary > 60000')

df['Department'] = ['HR','Finance','IT',' HR']

filtered_str = df[df['Department'].str.contains('H')]
p(filtered_str)

filtered_start = df[df['Name'].str.startswith('A')]
p(filtered_start)

filtered_end = df[df['Name'].str.endswith('d')]
p(filtered_end)

df['JoinDate'] = pd.to_datetime(['2022-01-01','2021-06-15','2023-03-10','2020-09-25'])
filtered_year = df[df['JoinDate'].dt.year == 2022]
p(filtered_year)

filtered_date_range = df[(df['JoinDate'] >= '2021-01-01') & (df['JoinDate'] <= '2022-12-31') ]
p(filtered_date_range)

filtered_boolean = df[(df['Age'] > 25)&(df['Salary'] > 60000 )]
p(filtered_boolean)

filtered_not = df[~(df['Age'] > 30)]
p(filtered_not)

# & and,  | or,  ~ not operadores de condición.
```

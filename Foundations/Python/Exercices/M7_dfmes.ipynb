````python
import pandas as pd
````

````python
data = pd.read_csv("hospitals.csv",sep = ";", encoding = "latin-1" )
````

### 2.Mostra totes les dades dels hospitals que tenen entre 100 i 120 "num_camas".

````python
print( data[data['num_camas'].between(100,120)] )
````

### 3.Mostra els noms dels hospitals que estan a Barcelona - Catalunya.

````python
print(data[ (data['comunidad_autonoma'] == 'Cataluña') & (data['localidad'] == 'Barcelona') ])
````

### 4.Mostra els 3 hospitals amb més "indice_satisfaccion" del grup d'hospitals de Madrid i Andalusia.

````python
print(data[data['comunidad_autonoma'].isin(['Comunidad de Madrid', 'Andalucia'])].sort_values('indice_satisfaccion',ascending=False).head(3))
````

###5.Mostra els hospitals que comencen per lletra "C" i terminen per lletra "a".

````python
print(data[data['nombre'].str.startswith('C') & data['nombre'].str.endswith('a')])
````

###6.Mostra el "indice_satisfaccion" mig de cada comunitat autònoma i província de la taula "hospitals".

````python
print( data.groupby(['comunidad_autonoma','provincia'])['indice_satisfaccion'].mean().round(2) )
````

###7.Mostra el "num_camas" total de cada comunitat autònoma.

````python
print( data.groupby('comunidad_autonoma')['num_camas'].sum() )
````

###8.Mostra el "porcentaje_ocupacion" més petit de cada província de cada comunitat autònoma.

````python
print( data.groupby(['comunidad_autonoma','provincia'])['porcentaje_ocupacion'].min() )
````

###9.Mostra quantes províncies i localitats té cada comunitat autònoma.

````python
print( data.groupby('comunidad_autonoma').agg(
    num_provinces=('provincia', 'nunique'),
    num_localidades=('localidad', 'nunique')
))
````

###10.Mostra les comunitats autònomes que tenen menys de 5 hospitals.

````python
hospital_counts = data.groupby('comunidad_autonoma')['hospital_id'].count()
print( hospital_counts[hospital_counts < 5].index.tolist() )
````

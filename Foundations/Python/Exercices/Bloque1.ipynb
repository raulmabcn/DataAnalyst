###1️⃣ Udemy – Análisis combinado

Muestra los cursos cuyo rating sea superior a la media de su propia categoría.

Obtén el curso con más subscribers de cada instructional_level.

Calcula el porcentaje de cursos gratuitos (is_paid = False) por categoría.

Muestra las categorías donde la mediana de reviews sea superior a 10.000.

```python
import pandas as pd
data = pd.read_csv('udemy_courses.csv')
print(data.info())
```

```python
print( data[ data['rating'] > data.groupby('category')['rating'].transform('mean')][['category', 'title', 'rating']] )

#category_mean = ( data.groupby('category', as_index = False ).agg( category_mean = ('rating','mean')))
#data_merge = data.merge(category_mean, on = 'category')
#print( data_merge[data_merge['rating'] > data_merge['category_mean']][['category','title','rating','category_mean']] )
```

```python
print( data[ data['num_subscribers'] ==  data.groupby('instructional_level')['num_subscribers'].transform('max')][['instructional_level','title','num_subscribers']])
#idx = data.groupby('instructional_level')['num_subscribers'].idxmax()
#print( data.loc[idx][['category','title','rating']] )
```

```python
data_paid = (~data['is_paid']).groupby(data['category']).mean() * 100
print(data_paid)
```

```python
data_category = data.groupby('category')['num_reviews'].median()
print( data_category[ data_category > 100 ] )
#data_category = data.groupby('category', as_index = False)['num_reviews'].median()
#print( data_category[ data_category['num_reviews'] > 100 ] )
```

###2️⃣ Hospitals – Filtros y métricas derivadas

Calcula una nueva columna ratio_pacientes_por_cama = total_pacientes / num_camas
(suponiendo que debes calcular primero los pacientes por hospital desde pacientes.csv).

Muestra los hospitales cuyo indice_satisfaccion esté por encima de la media de su comunidad autónoma.

Encuentra las provincias donde todos los hospitales tienen ocupación superior al 60%.

Muestra las comunidades autónomas donde el hospital con más camas tenga también un índice de satisfacción inferior a la media nacional.

```python
data_h = pd.read_csv('hospitals.csv', sep=';', encoding='latin-1')
data_p = pd.read_csv('pacientes.csv', sep=';', encoding='latin-1')
data_h.info()
data_p.info()
```

```python
data_p_h = data_p.groupby('hospital_id', as_index=False)['paciente_id'].count().rename(columns={'paciente_id':'total_pacientes'})
data_h_p = data_h.merge( data_p_h, on ='hospital_id' )
data_h['ratio_pacietes_por_cama'] = data_h_p['total_pacientes']/data_h_p['num_camas']
data_h
```

```python
data_h[data_h['indice_satisfaccion'] > data_h.groupby('comunidad_autonoma')['indice_satisfaccion'].transform('mean') ]['nombre']
```

```python
#( data_h['porcentaje_ocupacion'] > 60 ).groupby(data_h['provincia']).mean() == 1.0
provincias_validas = ( data_h['porcentaje_ocupacion'] > 60 ).groupby(data_h['provincia']).all()
provincias_validas[provincias_validas]
```

```python
national_mean =  data_h['indice_satisfaccion'].mean()
hospitals = data_h[(data_h.groupby(['comunidad_autonoma'])['num_camas'].transform('max') == data_h['num_camas']) ]
hospitals[hospitals['indice_satisfaccion'] < national_mean]['comunidad_autonoma'].unique()
```

```python
data_h.loc[
    (data_h['num_camas'] == data_h.groupby('comunidad_autonoma')['num_camas'].transform('max')) &
    (data_h['indice_satisfaccion'] < data_h['indice_satisfaccion'].mean()),
    'comunidad_autonoma'
].unique()
```

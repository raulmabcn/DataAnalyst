### 1. Exploració de dades del Udemy_courses.csv:

```python
import pandas as pd
df = pd.read_csv('udemy_courses.csv')
```

o Quantes columnes té el CSV, i quants registres?

```python
rows, columns = df.shape
print(f"{columns=}\n{rows=}")
```

o Quantes columnes de tipus enter té?

```python
print(f"Número de columnas de tipo entero: {df.select_dtypes( include = 'int64' ).shape[1]}" )
```

o I quantes de tipus decimal?

```python
print(f"Número de columnas de tipo decimal: {df.select_dtypes( include = 'float64' ).shape[1]}" )
```

o Quin és el nombre de reviews màxim? I el de subcribers?

```python
print(f"Número máximo de opiniones: {df['num_reviews'].max()}")
print(f"Número máximo de subscriptores: {df['num_subscribers'].max()}")
```

o Quina és la mitjana del ràting?

```python
print(f"Mediana de rating: {df['rating'].mean().round(2)}")
```

### 2. Filtres:

Selecciona els cursos que tenen més de 100.000 subscriptors, més d’un 4,7 de valoració i més de 90.000 ressenyes.

```python
print(f"Cursos seleccionados por boolean: {df[ ( df["num_subscribers"] > 100000 ) & (df["rating"] > 4.7) & (df["num_reviews"] > 90000 ) ]}")
print(f"Cursos selecionados por query: {df.query( "num_subscribers > 100000 and rating > 4.7 and num_reviews > 90000" )}")
```

Selecciona els cursos que tenen un "instructional level" de principiants i que són de la categoria "Development".

```python
print(f"Cursos seleccionados por boolean: {df[ (df["instructional_level"] == "Beginner Level") & (df["category"] == "Development") ]}")
print(f"Cursos seleccionados por query: {df.query( "instructional_level == 'Beginner Level' and category == 'Development'" )}")
```

### 3.Agrupacions

Mostra el nombre de cursos per categoria.

```python
print(f"Número de cursos por : {df.groupby('category')['id'].count()}")
```

Mostra la mitjana de valoració per "instructional level" i “is_paid”.

```python
print( f"Media del rating por :  {df.groupby(['instructional_level','is_paid'])['rating'].mean()}" )
```

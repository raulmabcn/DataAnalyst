### 1. Exploració de dades del Udemy_courses.csv:

```python
import pandas as pd
df = pd.read_csv('udemy_courses.csv')
```

o Quantes columnes té el CSV, i quants registres?

```python
rows, columns = df.shape
print(f"{columns=}\n{rows=}")
```

o Quantes columnes de tipus enter té?

```python
print(f"Número de columnas de tipo entero: {df.select_dtypes( include = 'int64' ).shape[1]}" )
```

o I quantes de tipus decimal?

```python
print(f"Número de columnas de tipo decimal: {df.select_dtypes( include = 'float64' ).shape[1]}" )
```

o Quin és el nombre de reviews màxim? I el de subcribers?

```python
print(f"Número máximo de opiniones: {df['num_reviews'].max()}")
print(f"Número máximo de subscriptores: {df['num_subscribers'].max()}")
```

o Quina és la mitjana del ràting?

```python
print(f"Mediana de rating: {df['rating'].mean().round(2)}")
```
